```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 2*np.pi, 10)
```

```python
len(x)
```

```python
x[:5]
```

```python
y = np.sin(x)
```

```python
y[:5]
```

```python
plt.figure(figsize = (6,4))
plt.plot(x, y, "x-", ms = 15, label = "$\sin(x)$")
plt.plot(x, np.cos(x), "r*-", color = "magenta", label = "$\cos(x)$")
plt.plot(x, np.sin(x) + np.cos(x), "D-.", ms = 6, color = "crimson", label = "suma")
plt.legend()
plt.xlabel("$x$", fontsize = 18)
plt.ylabel("$f$", fontsize = 18)
plt.title("Funciones", fontsize = 18)
plt.show()
```
